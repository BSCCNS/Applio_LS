```python
import numpy as np
import pandas as pd

import datamapplot
```

```python
ls_file = '/Users/tomasandrade/Documents/BSC/ICHOIR/applio/Applio_LS/experiments/songs_v1/feat_2d/feat_2d_layer_12.csv'
```

```python
df = pd.read_csv(ls_file, index_col=0)
df['start_min'] = df['start'].apply(lambda x: f"{int(x // 60):02d}:{x % 60:05.2f}")

df['label'] = df['song'] + ' |  ' + df['start_min']
mask = df['phone_base'] != 'SP'

mask2 = (df['phone_base'] == 'AP') & (df['duration'] > 0.3)

df_filter = df[mask & ~mask2]
df_filter
```

```python
plot = datamapplot.create_interactive_plot(
    df_filter[['x', 'y']].values,
    df_filter['phone_base'].values,
    # arxivml_label_layers[2],
    # arxivml_label_layers[4],
    hover_text = df_filter['label'].values,
    font_family="Playfair Display SC",
    enable_search=True
)
plot.save("my_plot.html")
```
