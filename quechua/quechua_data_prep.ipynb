```python
import glob
import pandas as pd
import os
from pathlib import Path
import shutil
```

```python
root = Path('/Users/tomasandrade/Documents/BSC/ICHOIR/datasets/IWSLT2024_Quechua_data')
```

```python
out_dir = root / 'wav'
```

```python
os.makedirs(out_dir, exist_ok=True)
```

```python
for dir in ['test', 'train', 'valid']:
    folder = root / 'que_spa_constrained' / dir / 'wav' 
    files = glob.glob(f'{folder}/*.wav')
    for f in files:
        file = Path(f)
        shutil.copy2(file, out_dir/ file.name)
```
