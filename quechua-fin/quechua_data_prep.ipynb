```python
import glob
import pandas as pd
import os
from pathlib import Path
import shutil

import soundfile as sf
```

```python
def get_duration_audio(path):
    info = sf.info(path)
    return info.duration
```

```python
root = Path('/Users/tomasandrade/Documents/BSC/ICHOIR/datasets/IWSLT2024_Quechua_data')
```

```python
out_dir = root / 'wav'
```

```python
os.makedirs(out_dir, exist_ok=True)
```

```python
for dir in ['test', 'train', 'valid']:
    folder = root / 'que_spa_constrained' / dir / 'wav' 
    files = glob.glob(f'{folder}/*.wav')
    for f in files:
        file = Path(f)
        shutil.copy2(file, out_dir/ file.name)
```

```python
root_fin = Path('/Users/tomasandrade/Documents/BSC/ICHOIR/datasets/MozillaCommonVoice-FI-2026-06-12/fi')
```

```python
def make_fin_df(path):
    df = pd.read_csv(path, sep="\t", encoding="utf-8", quoting=3)
    df_fin = df[['path', 'age', 'gender']].copy()
    df_fin['full_path'] = root_fin / 'clips' / df_fin['path']
    df_fin['duration'] = df_fin['full_path'].apply(get_duration_audio)

    return df_fin
```

```python
train_txt_path = root_fin / 'train.tsv'
test_txt_path = root_fin / 'test.tsv'

df_train = make_fin_df(train_txt_path).sample(frac= 0.7, random_state=1)
df_test  = make_fin_df(test_txt_path).sample(frac= 0.7, random_state=1)
```

```python
df_fin = pd.concat([df_train, df_test])
```

```python
df_fin['duration'].sum()/3600 
```
